```python
import numpy as np
import pandas as pd

df = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
df
```

# 1
Convert the column ‘TotalCharges’ to a numeric format, coercing errors if
necessary.


```python
df.dtypes['TotalCharges']
# O means object
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df.dtypes['TotalCharges']
```

#2

Change all column names to lowercase.

```python
df.columns = df.columns.str.lower()
df.columns
```

```python
df
```

#3

For string values change spaces to an underscore symbol.

```python
df = df.replace(' ','_', regex=True)
df
```

#4

Convert the column ‘churn’ to integer format (no - 0, yes - 1).

```python
df['churn'] = df['churn'].replace('No','0', regex=True)
df['churn'] = df['churn'].replace('Yes','1', regex=True)
df['churn'] = pd.to_numeric(df['churn'], errors='coerce')
print(df.dtypes['churn'])
df
```

Check for categorical columns how many different values they have.
